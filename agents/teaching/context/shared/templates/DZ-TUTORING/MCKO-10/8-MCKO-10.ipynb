№1 В алгоритме кодирования используется уравнение:

(d * e) % g(n) = 1

Функция  g(n)  вычисляется по формуле:

g(n) = (p-2)*(q-2)

Определите наибольшее значение числа  d , которое меньше 50, если известно, что  p = 7 ,  q = 11 ,  e = 13 .

№2 В некотором криптографическом алгоритме выполняется выражение:


(d * e) % h(n) = 1


Функция  h(n)  вычисляется по формуле:


h(n) = (p + 1) * (q + 1)


Определите наибольшее значение числа  d , которое меньше 60, если известно, что  p = 3 ,  q = 5 ,  e = 17 .

№3 Для кодирования применяется уравнение:


(d * e) % k(n) = 1


Где:


k(n) = (p - 3) * (q - 1)


Найдите наибольшее значение числа  d , которое меньше 45, если  p = 8 ,  q = 10 ,  e = 7 .
