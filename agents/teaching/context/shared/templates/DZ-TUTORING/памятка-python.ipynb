#### Строки

Строка (string | str) - включает в себя неогранниченное кол-во символов (это могут быть как буквы и цифры, так и любые символы $%^...) и обязательно обрамляется единичными или двойными кавычками ("Привет 123?*)

Со строками мы можем производить действия сложения и умножения

```'a' + 'b' = 'ab'```,```'a' * 3 = 'aaa'```

Для перевода число -> строка испльзуется str()

```str(10)```

Также мы можем измерить длину строки (посчитать кол-во символов):

```s = 'hello'```, ```len(s)``` ->``` 5```

Также мы можем посчитать кол-во определенных элементов в cтроке с помощью count:

```s = 'hello'```,``` s.count('l') ```->``` 2```

Также мы можем сортировать строки в порядок возростания (в случаями с числами) и в алфовитный порядок (в случае с остальными символами)

```s = 'bac'```, ```sorted(s) ```-> ```'abc'```



##### Числа

Число (number | int) - включает в себя неогранниченное кол-во цифр, которые для удобства можно разделять символами "_" через каждые три цифры (123_456_769).
!!Если число записанно в кавычках ("10"), то это уже строчный формат данных

Для перевода строка -> число испльзуется int() -> 

```int('10')```

- *(умножение) 4 * 3 = 12

##### Массивы

Массив | список (massive | list) - включает в себя неогранниченное кол-во элементов (это могут быть как строки, так и числа) записанных друг за другом через запятую ([1, 2, 'a', 'b'])
В массив мы можем можем добавлять элементы с помощью метода append:

```m = []```, ```m.append(1)```, ```m.append('heelo')```

Также мы можем измерить длину массива с помощью len (посчитать кол-во элементов):

```m = ['hello', 34, '4', 3]```,``` len(m)``` -> ```4```

Также мы можем посчитать кол-во определенных элементов в массиве с помощью count:

```m = [1, 1, 4, 3, 1]```,``` m.count(1)``` -> ```3```

Также мы можем сортировать массивы в порядок возростания (в случаями с числами) и в алфовитный порядок (в случае со строками)

```m = [3, 1, 2]```,``` sorted(m)``` -> ```[1, 2, 3]```

Также мы можем "склеить" массив в строку с помощью метода join:

```m = [1, 2, 3]```,``` "".join(m)``` -> ```[123]```

#### max | min

**Функции max() и min()**
В эти функции мы можем передать несколько аргументов (если это числа):

```min(1, 3, 2)``` -> ```3```

Также мы можем передать один аргумент (если это массив):

```m = [3, 2, 1, 4]```,``` max(m) ```-> ```4```

Также мы можем "скеить" массив в строку методом split:

```m = ['a', 'b', 'c']```, ```"".split(m)``` -> ```'abc'```

### Основы индексации

- *Индексация строк и массивов*

- *Обратная индексация ([-1])*

- *Обрезка строк и массивов*

**Индексами** в python называют индетифекаторы определенногых символов (в случае со строкой) или элементов (в случае с массивом), по которомым можно узнать порядковый номер этого символа/элемента в строке/массиве. Индекс записывается сразу после переменной в "[]", и начинается с 0 (порядковый номер первого элемента в массиве - 0:

`m = [1, 5, 7, 9]`, `m[1]` -> `5`, `m[2]` -> `7`

**Обратная индексация** (индексация с конца) - это индексация начиная с последнего элемента, который всегда имеет индекс [-1], например если мы не знаем точное кол-во букв в строке, но нам нужно получить из этой строки предпоследнюю букву:

`s = 'abc....zyx'`, `s[-2]` -> `y`

**Обрезка строк и массивов** - способ работы со строками, при котором мы можем обрезать строку/массив до нужных нам размеров и по необходимыи правилам. Например если мы хотим обрезать первые 2 символа в нашей строке, мы пишем нашу переменную строки, далее индекс в формате [n], и в нутри индекса ставим двоеточее для обрезки, по следуюшим правилам:

[`число с какого идекса начать обрезать`:`число ДО какого идекса обрезать`], по умолчанию левая часть равно нулю, а правая максимальному индексу в строке/массиве, тоесть если мы напишем:

`s = 'abcd'`, `s[2:]` -> то обрезка начнется с 0 идекса (так как мы не указали в ручную) и дойдем до 2го символа, и по итогу обрежет первые два символа строки -> `cd`

`s = 'abcd'`, `s[:2]` -> обрезка начнется со второго символа и дойдем до конца строки, что в итоге обрежет все символы строки, кроме первых двух -> `ab`

`s = 'abcd'`, `s[1:3]` -> обрезка начнется с 1 индекса включительно (в нашей строке это `d`) и закончится на 3 индексе не включительно (обрезка удалить все символы внутри строки, начиная со 2го и заканчивая 3им по счету элементами)  -> `ad`
