№1 На числовой прямой даны два отрезка: P  =  [10, 29] и Q  =  [13, 18].
Укажите наибольшую возможную длину отрезка A, для которого выражение

((x ∈ A) → (x ∈ P)) ∨ (x ∈ Q)

тождественно истинно, то есть принимает значение 1 при любом значении переменной х.

№2 На числовой прямой даны два отрезка: P  =  [69; 91] и Q  =  [77; 114]. Укажите наименьшую возможную длину такого отрезка A, для которого формула

(x ∈ Q) → (((x ∈ P) ≡ (x ∈ Q)) ∨ (¬(x ∈ P) → (x ∈ A)))

тождественно истинна (то есть принимает значение 1 при любом значении переменной х).


№3 На числовой прямой даны три отрезка: P  =  [13; 31], Q [18; 80] и R  =  [48; 114].

¬((x ∈ Q) → ((x ∈ P)∨(x ∈ R))) → (¬(x ∈ A) → ¬(x ∈ Q))

тождественно истинна (то есть принимает значение 1 при любом значении переменной х).
