### Основы

Начнем с оновой лексики, которую мы будем использовать на наших уроках, изначально теоретической информации будет очень много, но каждый элемент из этого списка мы еще много раз будем разбирать на практике.

**Переменная** - базовая еденица лексики любого ЯП, объявляется через =, название переменной может быть любым (желательно использовать понятные названия на латинице, например если переменная это число, то можно назвать ее n):

```n = 10*```

**Функция** - сложная единица/структура кода, имеет название (в основном на латинице) и скобочки после него:

```max()```

В скобочки функции мы передаем (записываем) **Аргументы** - переменные или любые данные, которые записываются в скобочках у функции через запятую:

```max(1, n, 'hello')```

**Методы** - те же функции, но с немного видоизмененной записи, сначала мы пишем объект к которому будем применять метод (например переменную), потом через "." записываем сам метод и сразу за ним круглые скобки, в которые можем передать **аргументы** (как и в функции):

```m.append('word')```

**Ввод/вывод информации** - для вывода информации в консоль мы используем функцию ptint(), в скобки которой мы передаем информацию, которую хотим вывести (строки, числа переменные и т.д.):

```a = 'hello!'```, ```print(s)``` -> ```hello!```
